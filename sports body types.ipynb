#Body type analysis in professional sports
##Background
Common to all team sports is the concept of "postion". In some sports, different positions differ mainly in high-level tactics, while in others, different positions may peform radically different actions. To the casual eye, these discrepancies seem to manifest in degree of body difference among players of a single sport. Football (as in not soccer), has one of the most extreme notions of position, and has clear body type segmentation (look at running backs vs. linemen). We can go beyond qualitative analysis using data and some machine learning. For instance which sport has the most defined body types?

```python
from bs4 import BeautifulSoup
from urllib2 import urlopen
import sys, re, pandas as pd, IPython, seaborn as sns, matplotlib.pyplot as plt, numpy as np, sklearn, colorsys
%matplotlib inline
sns.set_context(rc={'lines.markeredgewidth': 0.1}) #hack to make some seaborn plots appear correctly in IPython
```

##Getting the data
We'll scrape the data for the four major sports leagues in America (NBA, NFL, NHL, and MLB) from ESPN's website. (For convenience the data are also provided in csv files).

```python
#Web scraper for espn.com
def getLeagueStats(league):
    playerStats=[]
    teamLinks = getTeamLinks('http://espn.go.com/'+league.lower()+'/teams')
    for i in range(len(teamLinks)):
        print 'Examining '+league+ ' team ' + unicode(i+1) + ' of ' + unicode(len(teamLinks)) + '...'
        sys.stdout.flush()
        if league.lower()=='nhl': #nhl rosters have no position column
            playerStats.extend(getNHLTeamStats(teamLinks[i]))
        else:
            playerStats.extend(getTeamStats(teamLinks[i],league))
        IPython.display.clear_output(wait='True')
    print 'Done!'
    return playerStats

def getTeamLinks(leagueLink):
    leagueHTML = urlopen(leagueLink).read()
    leagueSoup = BeautifulSoup(leagueHTML,'lxml')
    teamLinks = leagueSoup.findAll('a',text='Roster')
    return ['http://espn.go.com' + tag['href'] for tag in teamLinks]

def getTeamStats(teamLink,league):
    (namecol,poscol,htcol,wtcol)=(1,2,4,5) #how espn's roster tables are arranged
    if league.lower()=='mlb': #baseball rosters are a little different
        (namecol,poscol,htcol,wtcol)=(1,2,6,7)
    teamHTML = urlopen(teamLink).read()
    teamSoup = BeautifulSoup(teamHTML,'lxml')
    teamTable = teamSoup.find('table')
    playerRows = [tag for tag in teamTable.children if (tag['class'] not in [['colhead'],['stathead'],['oddrow']])]
    playerStats = []
    for row in playerRows:
        stats = [tag.getText() for tag in row.children]
        height = int(stats[htcol].split('-')[0])*12 + int(stats[htcol].split('-')[1])
        playerStats.append({'name':stats[namecol], 'position': stats[poscol], 'height':height, 'weight':int(stats[wtcol])})
    return playerStats

def getNHLTeamStats(teamLink):
    (namecol,htcol,wtcol)=(1,3,4)
    posorder=['C','LW','RW','D','G'] #order of position groups on ESPN's website
    teamHTML = urlopen(teamLink).read()
    teamSoup = BeautifulSoup(teamHTML,'lxml')
    teamTable = teamSoup.find('table')
    playerRows = [tag for tag in teamTable.children]
    playerStats = []
    posnum=-1
    for row in playerRows:
        if row['class'][0]=='colhead':
            posnum+=1
        elif row['class'][0]=='stathead': pass
        else:
            stats = [tag.getText() for tag in row.children]
            height = int(stats[htcol].split('-')[0])*12 + int(stats[htcol].split('-')[1])
            playerStats.append({'name':stats[namecol], 'position': posorder[posnum], 'height':height, 'weight':int(stats[wtcol])})
    return playerStats 
```

```python
useScraper = True
if useScraper:
    nbaFrame = pd.DataFrame(getLeagueStats('NBA'))
    nflFrame = pd.DataFrame(getLeagueStats('NFL'))
    mlbFrame = pd.DataFrame(getLeagueStats('MLB'))
    nhlFrame = pd.DataFrame(getLeagueStats('NHL'))
    nbaFrame.to_csv('NBA stats.csv', encoding='utf-8')
    nflFrame.to_csv('NFL stats.csv', encoding='utf-8')
    mlbFrame.to_csv('MLB stats.csv', encoding='utf-8')
    nhlFrame.to_csv('NHL stats.csv', encoding='utf-8')
else:
    nbaFrame = pd.DataFrame.from_csv('NBA stats.csv')
    nflFrame = pd.DataFrame.from_csv('NFL stats.csv')
    mlbFrame = pd.DataFrame.from_csv('MLB stats.csv')
    nhlFrame = pd.DataFrame.from_csv('NHL stats.csv')
dfs={'NBA':nbaFrame,'NFL':nflFrame,'MLB':mlbFrame,'NHL':nhlFrame}
```

One thing that's immediately clear is the strange shape of the NFL's distribution. Anyone familiar with the sport would guess right away that the disjointed blob of data above 300 lbs is offensive and defensive linemen. Structure in the other sports is a little less obvious. We can study this more quantitatively by looking at how "clustered" the data points are. Specifically, we'll use K-Means clustering.
#Unsupervised learning to examine structure in body type distribution

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
from sklearn.cluster import KMeans
def fitKMeans(df,max_clusters=8):
    #preprocess data
    X = df[['height','weight']].values.astype(float)
    X_train, X_test = train_test_split(X, train_size=0.7, random_state=1)
    models=[]
    scores=[]
    clusters = np.arange(max_clusters)+1
    for i in clusters:
        newModel = Pipeline([('scaler',StandardScaler()),
                             ('clf',KMeans(n_clusters=i,random_state=1))])
        newModel.fit(X_train)
        models.append(newModel)
        #score method on Pipeline is broken for classifiers that don't take a y value, so code is duplicated/fixed here
        Xt=X_test
        for name,transform in newModel.steps[:-1]:
            Xt=transform.transform(Xt)
        #score on the test data
        scores.append(newModel.steps[-1][-1].score(Xt)/Xt.shape[0]) #divide by N_samples to normalize score
    return {'models':models, 'X train':X_train, 'X test':X_test, 'scores':scores, 'clust':clusters}
```

```python
nflKM = fitKMeans(nflFrame,10)
nbaKM = fitKMeans(nbaFrame,10)
nhlKM = fitKMeans(nhlFrame,10)
mlbKM = fitKMeans(mlbFrame,10)
```

Let's do a sanity check by plotting the 5 cluster case for the NFL test data.

```python
def getColors(N):
    #get N contrasting colors
    HSV = [(x*1.0/float(N), 1, 1) for x in range(N)]
    return [colorsys.hsv_to_rgb(*hsv) for hsv in HSV]
cdata = np.empty(nflKM['X test'].shape[0],dtype=('float,float,float'))
for i in nflKM['clust']:
    colors = getColors(i)
    cdata[nflKM['models'][4].predict(nflKM['X test'])==i-1]=colors[i-1]
plt.scatter(nflKM['X test'][:,0],nflKM['X test'][:,1],c=cdata.tolist(),alpha=0.3)
plt.xlabel('Height'); plt.ylabel('Weight')
```

Looks pretty good! Now lets look at how the clustering scores compare.

```python
KMs = [nflKM,nbaKM,nhlKM,mlbKM]
for k in KMs:
    plt.plot(k['clust'],k['scores'])
plt.legend(['NFL','NBA','NHL','MLB'],loc='lower right')
plt.xlabel('Clusters'); plt.ylabel('Score')
```

As one might expect, the NHL and MLB show less body type clustering than the NBA and NFL. What is surprising, though is that the NBA, despite the exclusive selection of extreme height as compared to the general public, shows at least as much clustering within it's exceptionally tall athletes as the NFL does! This is an interesting result, but there are some caveats about this that we'll discuss later.
##Supervised Learning to predict position
For now, lets consider the question of whether we can predict a player's position by height and weight. Based on the clustering results, I'd expect this to work well for the NBA and NFL and not so well for the MLB and NHL. First lets look at position data.

```python
for league,df in dfs.items():
    t=pd.pivot_table(df,index='position',aggfunc={'name':'count','height':np.mean,'weight':np.mean})[['name','height','weight']]
    t.columns=['num', 'avg ht', 'avg wt']
    print t
    print ""
```

We see a few positions with a very small number of players (usually a generic position like outfielder (OF)), so we'll remove them.

```python
nflFrame.drop(nflFrame['position'].where(nflFrame['position'].isin(['OL','NT','S'])).dropna().index,inplace=True)
mlbFrame.drop(mlbFrame['position'].where(mlbFrame['position']=='OF').dropna().index,inplace=True)
```

```python
sns.lmplot('height','weight',nbaFrame,hue='position',fit_reg=False)
plt.title('NBA')
g=sns.lmplot('height','weight',nhlFrame,hue='position',fit_reg=False)
plt.title('NHL')
```

Yep, the ordering by position is pretty obvious for the NBA and pretty non-obvious for the NHL. No surprises there. Let's use an SVM with a Gaussian kernel to learn more.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.grid_search import GridSearchCV
def fitSVM(df):
    #preprocess data
    X = df[['height','weight']].values.astype(float)
    y = df['position']
    X_train, X_test, y_train, y_test = sklearn.cross_validation.train_test_split(X, y, train_size=0.7, random_state=1)
    model = Pipeline([('scaler',StandardScaler()),
                      ('clf',SVC(kernel='rbf',random_state=1))])
    #find optimum C and gamma for the SVM using CV on the training set
    params = {'clf__gamma':np.logspace(-3,3,13),'clf__C':np.logspace(-3,3,13)}
    gscv=GridSearchCV(estimator=model,param_grid=params, verbose=1)
    gscv.fit(X_train,y_train)
    score = gscv.best_estimator_.score(X_test,y_test)
    return {'model':gscv.best_estimator_, 'X train':X_train, 'X test':X_test, 'y train':y_train, 'y test':y_test, 'gscv':gscv,
            'score':score}
```

```python
#This part'll take a few minutes on a slow computer
nflSVM=fitSVM(nflFrame)
nbaSVM=fitSVM(nbaFrame)
mlbSVM=fitSVM(mlbFrame)
nhlSVM=fitSVM(nhlFrame)
```

Let's check that our optimum SVM parameters aren't hitting the edge of the search grid

```python
svms = {'NBA':nbaSVM,'NFL':nflSVM,'MLB':mlbSVM,'NHL':nhlSVM}
for name,svm in svms.items():
    print "{:s}: C={:f} gamma={:f}".format(name,svm['model'].get_params()['clf__C'],svm['model'].get_params()['clf__gamma'])
```

So we could probably increase the C parameter for the NBA, but we won't sweat that right now. Great, and as a sanity check let's see what positions a 6', 180 lb player in the NBA and a 6'6" 350 lb player in the NFL should be playing

```python
print nbaSVM['model'].predict([72.0,180.0])[0]
print nflSVM['model'].predict([78.0,350.0])[0]
```

Looks good. Now let's look at the prediction scores

```python
for name,svm in svms.items():
    print "{}: {}".format(name,svm['score'])
```

So our NBA position predictor is 72% accurate, not bad. Unsurprisingly, our MLB and NHL predictors are not very good (though the MLB also has a high number of positions, making prediction much harder). It's something of a surprise that our NFL predictor is only 48% accurate, but again it's worth noting that the NFL has a large number of distinct positions. Let's drill down a little and see what's going on with our NFL data.

```python
svm=svms['NFL']
testdf=pd.DataFrame({'position':svm['y test'], 'prediction':svm['model'].predict(svm['X test'])})
testdf['correct']=(testdf['position']==testdf['prediction']).astype(int)
testdf.groupby('position').mean()
```

So we clearly see what's going on. Our SVM will only predict a subset of positions (likely due to some positions not being clustered and some positions having less players). We see some positions, like offensive tackle (OT) and linebacker (LB) are predicted well, while others like defensive end are not. We could tweak some things or try a different classifier, but there will always be a fundamental problem that there is no body type for which punter is the best guess at football position (because other positions have similar heights/weights and there are far fewer punters than, say, wide receivers). In contrast, let's do the same anaylsis for our relatively successful NBA SVM.

```python
svm=svms['NBA']
testdf=pd.DataFrame({'position':svm['y test'], 'prediction':svm['model'].predict(svm['X test'])})
testdf['correct']=(testdf['position']==testdf['prediction']).astype(int)
testdf.groupby('position').mean()
```

I'd call this a surprisingly good predictor given the minimal information we're using. Note that 87% NBA of point guards can be correctly identified by height and weight; that's pretty impressive.
##Caveats
This analysis is fun and has given us some surprising results about the degree of clustering within body types in the NBA, which to many peole just looks like a bunch of really tall guys. But there are some confounding factors. A huge one is that league reporting of heights and weights is likely to be skewed. Offensive linemen that in reality are over 400 lbs probably have their listed weight adjusted downward to be viewed as less "undisciplined", basketball centers who are 6'11" might be bumped to 7' for the cachet. Of course, it's hard to tell how much of this happens, and how much is actual selection of players with an "ideal" body. Another issue is that while height and weight are, in principle, continuous variables, they are reported discretely. This limits the precision of our analysis. Lastly, there's the issue of bias in the reporting of **position**. Take basketball for instance. LeBron James plays such a diverse array of roles on the court, he could probably be called any position (except center). He's likely listed as a small forward because of his size. Even in football, which has a very rigorous position system, there's a growing class of players like Jimmy Graham, Rob Gronkowski, Travis Kelce, etc. who effectively function in the offense as a wide receiver, but are listed as tight ends because they fit that physical mold.
With those qualifications, lets take a look at one last thing.
##Predicting sport from body type
Let's use an SVM again, but this time let's predict what sport someone plays based on their height and weight. We'll assemble our data into one big frame.

```python
nbaFrame['sport']='basketball'
nflFrame['sport']='football'
nhlFrame['sport']='hockey'
mlbFrame['sport']='baseball'
allFrame=pd.concat([nbaFrame,nflFrame,nhlFrame,mlbFrame])
```

```python
#This could take 20 minutes on a slow computer
X = allFrame[['height','weight']].values.astype(float)
y = allFrame['sport']
X_train, X_test, y_train, y_test = sklearn.cross_validation.train_test_split(X, y, train_size=0.7, random_state=1)
model = Pipeline([('scaler',StandardScaler()),
                  ('clf',SVC(kernel='rbf',random_state=1))])
#find optimum C and gamma for the SVM using CV on the training set
params = {'clf__gamma':np.logspace(-3,3,13),'clf__C':np.logspace(-3,3,13)}
gscv=GridSearchCV(estimator=model,param_grid=params, verbose=1)
gscv.fit(X_train,y_train)
score = gscv.best_estimator_.score(X_test,y_test)
model=gscv.best_estimator_
```

```python
score
```

About 62% prediction rate for sport based on body type. Let's break the predictions down by individual sport.

```python
testdf=pd.DataFrame({'sport':y_test, 'prediction':model.predict(X_test)})
testdf['correct']=(testdf['sport']==testdf['prediction']).astype(int)
testdf.groupby('sport').mean()
```

Despite the distinctive body type required by basketball, we only predict basketball players with a 66% success rate. On the other hand, we get 80% of football players right. Also, we again have a category (hockey) which isn't predicted at all. This makes sense given that the hockey body type is not very distinct, and the NHL has about half the athletes of MLB and a third the athletes of the NFL.

Let's plot the decision boundaries for our SVM. Because the Gaussian kernel SVM isn't easily "invertible" like some other algorithms, we'll just plot contours on a fine mesh.

```python
npts = 400 #400 points will look smooth but take a few seconds
xx,yy=np.meshgrid(np.linspace(X[:,0].min(),X[:,0].max(),num=npts),np.linspace(X[:,1].min(),X[:,1].max(),num=npts))
Zstring = model.predict(np.c_[xx.ravel(), yy.ravel()])
repdict = {'basketball':0,'football':1,'baseball':2,'hockey':3} #map predictions to plottable numbers
Z = np.zeros_like(Zstring)
for k,v in repdict.items():
    Z[Zstring==k]=v
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.get_cmap('Set2',4),vmin=0,vmax=3)
plt.xlabel('Height');plt.ylabel('Weight')
cbar = plt.colorbar(ticks=[0,1,2,3])
cbar.ax.set_yticklabels(['Basketball', 'Football', 'Baseball', 'Hockey'])
```

The NFL, having the most athletes of any of the sports we examined, occupies a big diagonal swath of correlated height and weight. Unsurprisingly, basketball dominates among taller bodies. Baseball is the best prediction for two body types: moderate height and low weight (most baseball players), and low height and large weight (Prince Fielder/Bartolo Colon types). This concludes our analysis; one obvious extension question is how these body type statistics have changed over time (besides the obvious fact that professional athletes have gotten bigger).
